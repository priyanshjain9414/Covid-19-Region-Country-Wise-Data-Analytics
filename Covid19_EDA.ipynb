This report presents an in-depth analysis of the global COVID-19 pandemic using
country-wise data grouped by WHO regions. The main goal is to understand how the
number of confirmed cases, deaths, recoveries, and active cases vary across different
parts of the world.

Through Exploratory Data Analysis (EDA), the study examines key indicators to
identify which regions were most affected and which managed to control the spread
more effectively. The dataset includes information from 187 countries, covering total
and weekly statistics.

Various data visualizations such as bar charts and pie charts are used to highlight
regional differences and patterns. These visuals help simplify complex data and provide
clear insights into the spread, impact, and recovery trends of COVID-19 across the
globe.

## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Dataset

```python
df = pd.read_csv('country_wise_Covid19.csv')
df.head()
```

```python
print("----- MISSING VALUES -----")
print(df.isnull().sum())
```

```python
print("----- DUPLICATES -----")
print(df.duplicated().sum())
```

## 3. Global Overview

```python
total_confirmed = df['Confirmed'].sum()
total_deaths = df['Deaths'].sum()
total_recovered = df['Recovered'].sum()
total_active = df['Active'].sum()

global_death_rate = (total_deaths / total_confirmed) * 100
global_recovery_rate = (total_recovered / total_confirmed) * 100
global_active_rate = (total_active / total_confirmed) * 100

print("Global Totals:")
print(f"Confirmed: {total_confirmed:,}")
print(f"Deaths: {total_deaths:,}")
print(f"Recovered: {total_recovered:,}")
print(f"Active: {total_active:,}")
print(f"Global Death Rate: {global_death_rate:.2f}%")
print(f"Global Recovery Rate: {global_recovery_rate:.2f}%")
```

### Pie chart for global case distribution

```python
plt.figure(figsize=(6,6))
plt.pie([total_active, total_recovered, total_deaths],
        labels=['Active','Recovered','Deaths'],
        autopct='%.1f%%', startangle=120)
plt.title("Global COVID-19 Case Distribution")
plt.show()
```

## 4. WHO Region-wise Analysis

```python
region_summary = df.groupby('WHO Region')[['Confirmed','Deaths','Recovered','Active','1 week % increase']].agg(['sum','mean']).reset_index()
region_summary.columns = ['WHO Region', 'Total Confirmed', 'Avg Confirmed', 'Total Deaths', 'Avg Deaths',
                          'Total Recovered', 'Avg Recovered', 'Total Active', 'Avg Active',
                          'Total Weekly %', 'Avg Weekly %']

print(region_summary)
```

### Bar plot: Confirmed cases by WHO Region

```python
plt.figure(figsize=(10,6))
sns.barplot(x='WHO Region', y='Total Confirmed', data=region_summary, palette='coolwarm')
plt.title("Total Confirmed Cases by WHO Region")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Grouped bar chart: Recovered vs Deaths per region

```python
plt.figure(figsize=(10,6))
region_bar = df.groupby('WHO Region')[['Recovered','Deaths']].sum().reset_index()
region_bar.plot(kind='bar', x='WHO Region')
plt.title("Recovered vs Deaths per WHO Region")
plt.xticks(rotation=60)
plt.show()
```

### Pie chart: Deaths by Region

```python
plt.figure(figsize=(6,6))
plt.pie(region_summary['Total Deaths'], labels=region_summary['WHO Region'], autopct='%.1f%%', startangle=140)
plt.title("Share of Deaths by WHO Region")
plt.show()
```

## 5. Top 10 Country Analysis

```python
top10_confirmed = df.nlargest(10, 'Confirmed')
top10_deaths = df.nlargest(10, 'Deaths')
top10_active = df.nlargest(10, 'Active')
top10_growth = df.nlargest(10, '1 week % increase')
```

### Top 10 Most Affected Countries

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Country/Region', y='Confirmed', data=top10_confirmed, palette='Blues_d')
plt.title("Top 10 Countries by Confirmed Cases")
plt.xticks(rotation=45)
plt.show()
```

### Top 10 Countries Having High Deaths

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Country/Region', y='Deaths', data=top10_deaths, palette='Reds_d')
plt.title("Top 10 Countries by Deaths")
plt.xticks(rotation=45)
plt.show()
```

### Top 10 Countries Having High Active Cases

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Country/Region', y='Active', data=top10_active, palette='Greens_d')
plt.title("Top 10 Countries by Active Cases")
plt.xticks(rotation=45)
plt.show()
```

### Top 10 Fast Spread Countries

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Country/Region', y='1 week % increase', data=top10_growth, palette='Purples')
plt.title("Top 10 Countries by Weekly % Increase")
plt.xticks(rotation=45)
plt.show()
```

## 6. Weekly Change Analysis

```python
plt.figure(figsize=(12,6))
sns.histplot(df['1 week % increase'], bins=25, kde=True, color='orange')
plt.title("Distribution of 1 Week % Increase in Cases")
plt.xlabel("1 Week % Increase")
plt.ylabel("Number of Countries")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='WHO Region', y='1 week % increase', data=df, palette='cool')
plt.title("Weekly % Increase Distribution by WHO Region")
plt.xticks(rotation=45)
plt.show()
```

```python
fast_growing = df[df['1 week % increase'] > 20][['Country/Region','1 week % increase','WHO Region']]
print("Countries with >20% Weekly Increase:")
print(fast_growing)
```

## 7. Correlation Analysis

```python
corr = df[['Confirmed','Deaths','Recovered','Active','1 week % increase',
           'Deaths / 100 Cases','Recovered / 100 Cases']].corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation Heatmap of Major COVID-19 Metrics")
plt.show()
```

## 8. Relationship Plots

### Confirmed vs Deaths

```python
sns.scatterplot(x='Confirmed', y='Deaths', data=df, hue='WHO Region', s=80, alpha=0.8)
plt.title("Confirmed vs Deaths by WHO Region")
plt.show()
```

### Confirmed vs Recovered

```python
sns.scatterplot(x='Confirmed', y='Recovered', data=df, hue='WHO Region', s=80, alpha=0.8)
plt.title("Confirmed vs Recovered by WHO Region")
plt.show()
```

```python
sns.scatterplot(x='Deaths / 100 Cases', y='Recovered / 100 Cases', hue='WHO Region', data=df, s=70)
plt.title("Death Rate vs Recovery Rate by WHO Region")
plt.show()
```

## 9. Mortality and Recovery Rate

```python
df['Mortality Rate (%)'] = (df['Deaths'] / df['Confirmed']) * 100
df['Recovery Rate (%)'] = (df['Recovered'] / df['Confirmed']) * 100
```

### Top 15 countries by mortality rate

```python
top15_mortality = df.nlargest(15, 'Mortality Rate (%)')
plt.figure(figsize=(12,6))
sns.barplot(x='Country/Region', y='Mortality Rate (%)', data=top15_mortality, palette='Reds')
plt.title("Top 15 Countries by Mortality Rate (%)")
plt.xticks(rotation=60)
plt.show()
```

### Top 15 countries by Recovery rate

```python
top15_mortality = df.nlargest(15, 'Recovery Rate (%)')
plt.figure(figsize=(12,6))
sns.barplot(x='Country/Region', y='Recovery Rate (%)', data=top15_mortality, palette='Reds')
plt.title("Top 15 Countries by Recovery Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

### Violin plot for Recovery Rate by WHO Region

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='WHO Region', y='Recovery Rate (%)', data=df, palette='Set3')
plt.title("Distribution of Recovery Rate by WHO Region")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df[['Confirmed','Deaths','Recovered','Active','1 week % increase']], diag_kind='kde')
plt.suptitle("Pairwise Relationship Between Key COVID-19 Metrics", y=1.02)
plt.show()
```
